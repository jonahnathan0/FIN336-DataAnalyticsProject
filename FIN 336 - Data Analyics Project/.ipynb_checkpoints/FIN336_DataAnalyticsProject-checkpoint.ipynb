```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols as sm_ols
from statsmodels.iolib.summary2 import summary_col
import statsmodels.api as sm
```

```python
df = pd.read_excel("Spring2025_Freddie_Assignment_Data.xlsx")
```

```python
df
```

```python
column_rename_map = {
    'v1': 'fico',
    'v2': 'first_payment_date',
    'v3': 'first_time_homebuyer_flag',
    'v4': 'maturity_date',
    'v5': 'msa',
    'v6': 'mi_pct',
    'v7': 'num_units',
    'v8': 'occupancy_status',
    'v9': 'cltv',
    'v10': 'dti',
    'v11': 'orig_upb',
    'v12': 'ltv',
    'v13': 'interest_rate',
    'v14': 'channel',
    'v15': 'ppm_flag',
    'v16': 'amortization_type',
    'v17': 'property_state',
    'v18': 'property_type',
    'v19': 'postal_code',
    'v20': 'loan_seq_num',
    'v21': 'loan_purpose',
    'v22': 'orig_loan_term',
    'v23': 'num_borrowers',
    'v24': 'seller_name',
    'v25': 'servicer_name',
    'v26': 'super_conforming_flag',
    'v27': 'pre_relief_seq_num',
    'v28': 'program_indicator',
    'v29': 'relief_refinance_indicator',
    'v30': 'property_valuation_method',
    'v31': 'interest_only_indicator',
    'v32': 'mi_cancellation_indicator'
}

df = df.rename(columns=column_rename_map)
```

```python
df
```

```python
missing_value_codes = [9, 99, 999, 9999, '9', '99', '999', '9999']
df.replace(missing_value_codes, np.nan, inplace=True)
cols_to_numeric = ['fico', 'dti', 'ltv', 'cltv', 'interest_rate', 'orig_upb', 'orig_loan_term']
df[cols_to_numeric] = df[cols_to_numeric].apply(pd.to_numeric, errors='coerce')
df['first_payment_date'] = pd.to_datetime(df['first_payment_date'], format='%Y%m', errors='coerce')
df['maturity_date'] = pd.to_datetime(df['maturity_date'], format='%Y%m', errors='coerce')
```

```python
df
```

```python
df['first'] = (df['first_time_homebuyer_flag'] == 'Y').astype(int)
df['penalty'] = (df['ppm_flag'] == 'Y').astype(int)

cols = ['interest_rate', 'orig_upb', 'orig_loan_term', 'ltv', 'dti', 'fico', 'num_borrowers', 'first', 'penalty']
summary = df[cols]
stats = summary.agg(['mean', 'median', 'std', 'min', 'max', 'count']).T

decimals = {'interest_rate': 1, 'orig_upb': 0, 'orig_loan_term': 0, 'ltv': 1,
            'dti': 1, 'fico': 0, 'num_borrowers': 1, 'first': 2, 'penalty': 3}

stats_df = stats.copy().astype(str)
for var in stats_df.index:
    d = decimals[var]
    stats_df.loc[var] = stats.loc[var].apply(lambda x: f"{x:.{d}f}")

stats_df
```

```python
cols = ['interest_rate', 'orig_upb', 'orig_loan_term', 'ltv', 'dti',
        'fico', 'num_borrowers', 'first', 'penalty']

correlation_matrix = df[cols].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Pairwise Correlation Matrix')
plt.show()
```

```python
df['purchase'] = (df['loan_purpose'] == 'P').astype(int)
df['fico_under_680'] = (df['fico'] < 680).astype(int)
df['single_borrower'] = (df['num_borrowers'] == 1).astype(int)

group_cols = ['interest_rate', 'purchase', 'orig_upb', 'ltv', 'dti',
              'fico', 'fico_under_680', 'single_borrower']

yearly_avg = df.groupby('oyear')[group_cols].mean()
overall_avg = pd.DataFrame(yearly_avg.mean(axis=0)).T
overall_avg.index = ['Total']

df_avg = pd.concat([yearly_avg, overall_avg])

decimals = {'interest_rate': 1, 'purchase': 2, 'orig_upb': 0, 'ltv': 1,
            'dti': 1, 'fico': 0, 'fico_under_680': 2, 'single_borrower': 2}

df_avg_final = df_avg.copy()
for col in df_avg.columns:
    d = decimals[col]
    df_avg_final[col] = df_avg[col].round(d).astype(str)
df_avg_final
```

```python
df_seller = df['seller_name'].value_counts().reset_index()
df_seller.columns = ['Seller', 'Frequency']

df_seller['Percent'] = round((df_seller['Frequency'] / df_seller['Frequency'].sum()) * 100, 2)
df_seller['Cumulative'] = round(df_seller['Percent'].cumsum(), 2)

df_seller.insert(0, 'Number', [f"{i+1}." for i in range(len(df_seller))])
df_seller
```

```python
df_reg = df[(df['oyear'] >= 2013) & (df['oyear'] <= 2024)].copy()
df_reg = pd.get_dummies(df_reg, columns=['oyear'], drop_first=True)

reg1 = sm_ols(
    'interest_rate ~ orig_upb + orig_loan_term + ltv + dti + fico + '
    'oyear_2014 + oyear_2015 + oyear_2016 + oyear_2017 + oyear_2018 + '
    'oyear_2019 + oyear_2020 + oyear_2021 + oyear_2022 + oyear_2023 + oyear_2024',
    data=df_reg
).fit()


info_dict = {
    'R-squared': lambda x: f"{x.rsquared:.2f}",
    'Adj R-squared': lambda x: f"{x.rsquared_adj:.2f}",
    'No. observations': lambda x: f"{int(x.nobs)}"
}

print(summary_col([reg1], 
                  stars=True,
                  info_dict=info_dict))
```

```python
df_reg = df[(df['oyear'] >= 2013) & (df['oyear'] <= 2024)].copy()
df_reg = pd.get_dummies(df_reg, columns=['oyear'], drop_first=True)

X_cols = ['orig_upb', 'orig_loan_term', 'ltv', 'dti', 'fico'] + \
         [col for col in df_reg.columns if col.startswith('oyear_')]

df_reg = df_reg[X_cols + ['interest_rate']].dropna()

X = df_reg[X_cols].astype(float)
y = df_reg['interest_rate'].astype(float)

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
print(model.summary())
```
